# AI 개발 교육 과정을 시작한 계기
우연히 중진공(중소벤처기업진흥공단)에서 지원하는 AI 교육 과정에 대해 알게 되었다.

처음 알게 된 계기는 과거 유니티 프로그램을 공부하기 위해 구독한 유튜브 채널에서 소개한 영상 때문이었다.

AI에 대해 하나도 몰랐지만 어렴풋이 AI가 앞으로 더 널리 쓰이게 될 것이라는 막연한 생각만 있었다.

AI에 대해 처음 알게 된 것은 과거 게임기획자 교육을 받고 면접을 봤을 때였다.

지원을 잘못하여 AI 개발 직군으로 이력서를 넣게 되고 면접을 보게 되었는데, 그 때 AI 개발에 대해 처음 언급을 들었다.

당연히 면접에서 떨어졌고, 게임기획자 쪽 취업은 포기하게 되었다.

어쩌면 이때 면접에서 떨어진 것 때문에 AI에 흥미가 생긴 것인지도 모르겠다.

1차로 인공지능 상식 평가과와 적성검사를 하고 2차로 면접을 보았다.

1기 교육생들의 후기 등을 찾아보며 인공지능 상식 평가에 대해 알아보았고 예시 문제를 10문제 정도 보았는데 컴퓨터 관련해 무지한 나는 모르는 내용이 대부분이었다.

허겁지겁 2주 정도 인터넷을 찾아보며 공부를 하였고 1차 지원을 마쳤다. 

무사히 2차 면접을 보게 되었고 면접 준비를 한 후 면접을 봤지만 내가 준비한 것과는 다르게 대부분 나의 행적들과 앞으로 잘 할 수 있는지에 대한 질문이 많았다.

아마도 내가 컴퓨터 쪽 공부 경험이 아예 없기 때문인가 싶다. (내 앞의 지원자는 개발 쪽 질문을 엄청 하는 것이 들렸다.)

아무튼 2차 면접까지 마치고 결과 발표를 기다리는데 결과는 추가 합격이었다. 

대기 번호 10번이었는데 사실 별로 기대안하고 있었다가 2주일 후 추가 합격 통지를 받게 되었다.

다음날 바로 교육장에 갔는데 내 생각과는 완전 딴판이었다. 

오프라인 교육생 100명, 온라인 교육생을 100명 뽑는데 난 지식이 없어서 오프라인을 선택했다.

막상 교육을 보니 줌을 통해 수업을 진행하였고 온라인과 오프라인 교육생 모두 줌으로 교육을 받았다.

추가 합격으로 늦게 수업에 참여하다보니 기본적인 것들도 몰랐고 엄청 헤맸지만, 옆자리에 앉은 교육이 친절하게 알려주었고 교육장에 계신 매니저분들도 친절하게 알려주셔서 천천히 따라갈 수 있게 되었다.

교육장에 애플 컴퓨터인 아이맥이 있었는데 애플 기기는 한번도 써본 적 없어서 매우 어려웠고 처음 2일 동안은 아이맥으로 수업을 진행하다가 이대로는 도저히 수업을 못따라갈 것이라 생각해 결국 내 노트북을 갖고 수업을 듣기로 하였다. (아이맥을 쓰는 이유는 성능이 엄청 좋고 대부분 개발자들이 쓰기 때문이다.)
