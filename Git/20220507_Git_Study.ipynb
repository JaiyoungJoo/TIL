### Branch
깃허브에서 만든 레포 폴더 들어간 상태에서 시작

git branch : 브랜치 확인, 처음엔 main 만 있다.

git branch year-3 : year-3브랜치 만들기

git switch year-3 : 브랜치 변경

git puch origin year-3 : year-3 브랜치에서 작업한 것 올리기

(git config --global core.autocrlf true

이거 치면 add 했을 때 제대로 나옴)

### merge

다른 브랜치에서 작업한 것 가져오기

git switch main

git merge year-3 : 브랜치에서 작업한게 main에 그대로 있다.

git branch -D year-3 : 브랜치 지우기
